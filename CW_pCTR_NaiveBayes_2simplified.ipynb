```python
# Baseline model

def static_bid_val(df_data, bid, random_bid= False, print_YN= False):
    """
    """
    total_cost = 0
    budget = 250000
    imp_w = 0
    imp_l = 0
    clicks = 0
    lst_win = {}
    
    if random_bid== False:
        our_bid= bid
    else:
        max_bid= bid
        
    df_data_sub = df_data[['bidid', 'click', 'payprice']]

    for bidid, click, payprice in  df_data_sub.values:
        if random_bid== True:
            our_bid= random.randint(0, max_bid)
            
        if (total_cost + payprice) > budget:
            break
            
        if payprice < our_bid: #and click > 0
            imp_w += 1
            clicks += click
            total_cost += payprice
            lst_win[bidid]= [payprice, click]
            
        if payprice >= our_bid:
            imp_l += 1
            
    imp_total = len(df_data)
    try:
        imp_comp = round((imp_w+ imp_l)/ imp_total, 3)
    except: 
        imp_comp = 0
    try:
        bud_spent = total_cost/ budget
    except: 
        bud_spent = 0        
    try:
        CTR = round((clicks/ imp_w), 5)
    except:
        CTR = 0
    
    if print_YN== True:
        print('Bids won:', imp_w)
        print('Bids lost:', imp_l)
        print('% bids comp', imp_comp)
        print('Clicks:', clicks)
        print('CTR:', CTR)
        print('Total cost:', total_cost)
    
    return clicks, CTR, imp_w, imp_comp, bud_spent

clicks, CTR, imp_w, imp_comp, bud_spent= static_bid_val(df_val[:10000], 300, False, True)
```

```python
# Improved model with pCTR

def dynamic_model_val(df_data, baseline_bid, pCTR_data, print_YN='Yes'):
    total_cost = 0
    budget = 250000
    imp_w = 0
    imp_l = 0
    clicks = 0
    lst_win= []
    
    df_data_sub = df_data[['bidid', 'click', 'payprice']]
    
    counter = 0 
    for bidid, click, payprice in df_data_sub.values:
        if (total_cost + payprice) > budget:
            break
        
        our_bid = baseline_bid* pCTR_data[counter]
        counter += 1
        
        if payprice < our_bid: 
            imp_w += 1
            clicks += click
            total_cost += payprice
            
        if payprice >= our_bid:
            imp_l += 1
    
    imp_total = len(df_data)
    try:
        imp_comp = round((imp_w+ imp_l)/ imp_total, 3)
    except: 
        imp_comp = 0
    try:
        bud_spent = total_cost/ budget
    except: 
        bud_spent = 0        
    try:
        CTR = round((clicks/ imp_w), 5)
    except:
        CTR = 0
    
    if print_YN== 'Yes':
        print('Bids won:', imp_w)
        print('Bids lost:', imp_l)
        print('% bids comp', imp_comp)
        print('Clicks:', clicks)
        print('CTR:', CTR)
        print('Total cost:', total_cost)
    
    return clicks, CTR, imp_w, imp_comp, bud_spent
```
